```python
def solution(new_id):
    exception_mark = ['-', '_', '.']
    answer = ''
    answer_backup = ''
    # 1단계
    a = new_id.lower()
    print(a)

    # 2단계
    for i in a:
        if i.isalpha() == False and i.isdigit() == False:
            if i not in exception_mark:
                pass
            else:
                answer_backup += i
        else:
            answer_backup += i
    print(answer_backup)

    # 3단계
    for i in range(len(answer_backup)):
        if answer_backup[i] == '.':
            try:
                if answer_backup[i+1] == '.':
                    continue
                else:
                    answer += answer_backup[i]
            except IndexError:
                answer += answer_backup[i]
        else:
            answer += answer_backup[i]
    print(answer)

    # 4단계
    try:
        if answer[0] == '.':
            answer = answer[1:]
        if answer[-1] == '.':
            answer = answer[:-1]
    except IndexError:
        pass
    print(answer)

    # 5단계
    if len(answer) == 0:
        answer += 'a'
    print(answer)

    # 6단계
    if len(answer) >= 16:
        answer = answer[:15]
        while answer[-1] == '.':
            answer = answer[:-1]
    print(answer)

    # 7단계
    if len(answer) <= 2:
        while len(answer) != 3:
            answer += answer[-1]

    print (answer)
    
    return answer
```
